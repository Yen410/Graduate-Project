```python
import matplotlib.pyplot as plt
import numpy as np
import pandas
data = pandas.read_csv("/Users/yanting/health/project/d/d_age.csv",low_memory=False)
```

```python
data
```

```python
plt.rcParams['font.sans-serif'] =['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False
bar_width = 0.4
plt.bar(np.arange(7), data["men"], label = 'men', color = 'steelblue', alpha = 0.8, width = bar_width)
plt.bar(np.arange(7)+bar_width, data["women"], label = 'women', color = 'indianred', alpha = 0.8, width = bar_width)
plt.xticks(np.arange(7)+bar_width,data["time"])
plt.ylim([70, 85])
for x2016,y2016 in enumerate(data["men"]):
    plt.text(x2016-0.3, y2016, '%s' %y2016)

for x2017,y2017 in enumerate(data["women"]):
    plt.text(x2017+0.1, y2017, '%s' %y2017)
plt.title('n=101373')
#接受認知障礙功能檢測之歷年男女年齡比例
plt.xlabel('time')
plt.ylabel('age')
plt.legend()
plt.show()
```
